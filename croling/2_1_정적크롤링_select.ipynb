```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bs
```

```python
html = urlopen("https://www.nate.com")
soup = bs(html, "html.parser")
css_a = soup.select("#divGnb")[0].find_all("a") # 선택자 복자 
for a  in css_a:
    print(a.text)
```

```python
# 네이버 지식인에 문래역을 검색한 후에 제목만 갖고오세요  
html = urlopen("https://kin.naver.com/search/list.naver?query=%EB%AC%B8%EB%9E%98%EC%97%AD")
soup = bs(html, "html.parser")
css_a = soup.select("#s_content > div.section > ul")
```

```python
print(css_a[0].find_all("li")[0].find("dt").get_text()) # method
print(css_a[0].find_all("li")[0].find("dd").text) # value
print(css_a[0].select("#s_content > div.section > ul > li:nth-child(1) > dl > dd:nth-child(3)")[0].get_text()) # id, class 가 없으므로 selector를 사용한다. 
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
url = 'https://www.melon.com//genre/song_list.htm?gnrCode=GN0300'
table_raw = requests.get(url, headers = header)
# table_bs = BeautifulSoup(table_raw.text, "html.parser")
pd.read_html(table_raw.text)[0][["NO", "곡정보", "앨범"]]
```
