```python
# 변수 참고 사이트 (https://mazdah.tistory.com/883)

# SalePrice  - 자산의 판매 가격(달러)입니다. 이것이 여러분이 예측하려는 목표 변수입니다.
# MSSubClass: 주거유형
# MSZoning: 위치한 구역정보
# LotFrontage: 집에서 도로까지의 직선 거리 (feet 단위)
# LotArea: 부지 면적
# Street: 집까지 이어지는 길의 형태
# Alley: 집에 접근하는 골목의 형태
# LotShape: 집의 일반적인 형태
# Land Contour: 집의 평탄도
# Utilities: 사용 가능한 편의 시설 형태
# LotConfig: 부지 구성 (도로와의 인접 형태)
# LandSlope: 집의 경사
# Neighborhood: 도시 경계로부터의 물리적 위치
# Condition1: 집 주변의 다양한 환경 - 간선도로 및 철도 인접 여부
# Condition2:  집 주변의 다양한 환경 (13번에서 선택한 것 외에 추가적인 환경이 있는 경우 선택, 항목은 13과 동일)
# BldgType: 주거 유형 (단독 주택, 다세대 주택, 타운하우스 등)
# HouseStyle: 주택 형태 (층수 및 현관 구조)
# OverallQual: 전체 재료 및 마감 품질
# OverallCond: 집의 전반적인 상태
# YearBuilt: 시공일
# YearRemodAdd: 리모델링 날짜(리모델링 없는 경우 시공일과 동일)
# RoofStyle: 지붕형태
# RoofMatl: 지붕 재료
# Exterior1st: 집 외장
# Exterior2nd: 집 외장 (추가 선택 항목, 2가지 이상의 재료가 쓰인 경우 23번에 이어 추가로 선택)
# MasVnrType: 벽돌 외장 유형
# MasVnrArea: 벽돌 외장 면적
# ExterQual: 외장재 품질
# ExternCond: 외부 소재의 현재 상태
# Foundation: 건물 바닥재 종류
# BsmtQual: 지하실 높이(높이 값이 아니라 높이에 따른 좋고 나쁨의 정도)
# Bsmt Cond: 지하실의 일반적인 상태
# BsmtExposure: 지하실에서 밖으로 나가는 문이 있는지 혹은 벽 높이가 정원 수준인지
# BsmtFinType1: 지하실 마감 등급
# BsmtFinSF1: 첫 유형의 지하실 마감 면적 (평방피트)
# BsmtFinType2: 지하실 마감 등급(다중 형태인 경우)
# BsmtFinSF2: 두 번째 유형의 지하실 마감 면적 (평방피트)
# BsmtUnfSF: 마감되지 않은 지하실 면적 (평방피트)
# TotalBsmtSF: 지하실의 전체 면적 (평방피트)
# Heating: 난방의 종류
# HeatingQC: 난방 품질 및 상태
# CentralAir: 중앙 냉방 여부
# Electrical: 전기 시스템
# 1stFlrSF: 1층 넓이
# 2ndFlrSF: 2층 넓이
# LowQualFinSF: 전체 층 중 낮은 품질로 마감된 면적 (평방피트)
# GrLivArea: 주거용 지상 공간의 넓이 (평방피트)
# BsmtFullBath: 지하실의 (모든 시설이 갖춰진)욕실 수
# BsmtHalfBath: 지하실의 (샤워시설 및 욕조가 없는)욕실 수
# FullBath: 지상층의 (모든 시설이 갖춰진)욕실 수
# HalfBath: 지상층의 (샤워시설 및 욕조가 없는)욕실 수
# Bedroom: 지상층의 침실 수 (지하실의 침실은 제외)
# Kitchen : 지상층의 주방 개수
# KitchenQual : 주방퀄리티
# TotTrmsAbvGrd: 지상층의 전체 방 수 (욕실 제외)
# Functional: 집의 기능성 (공제가 보장되지 않는 경우 일반으로 가정) - 집이 제구실을 하는지?
# Fireplaces: 벽난로 개수
# FireplaceQu: 벽난로 품질
# GarageType: 차고 위치
# Garage YrBlt: 차고가 건설된 연도
# GarageFinish: 차고의 인테리어 마감
# GarageCars: 주차 가능 대수 기준 차고 크기
# GarageArea: 차고 면적 (평방피트)
# Garage Qual: 차고 품질
# GarageCond: 차고 상태
# PavedDrive: 진입로 포장 상태
# WoodDeckSF: 나무 바닥 면적 (평방피트)
# OpenPorchSF: 개방된 베란다 면적 (평방피트)
# EnclosedPorch: 막힌 베란다 면적 (평방피트)
# 3SsnPorch: 3계절 베란다 면적 (평방피트)
# ScreenPorch: 유리벽 베란다 면적 (평방피트)
# PoolArea: 풀장 면적 (평방피트)
# PoolQC: 풀 품질
# Fence: 울타리 품질
# MiscFeature: 기타 범주에서 다루지 않는 기타 기능
# MiscVal: $기타 기능의 값
# MoSold: 판매된 달
# YrSold: 판매 연도
# SaleType: 판매유형
# SaleCondition: 판매 상태
```
