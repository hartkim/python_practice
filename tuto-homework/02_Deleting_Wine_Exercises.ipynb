### Step 6. alcohol의 처음 3개 행의 값을 NaN으로 설정하세요

```python
# 보통 인덱싱, 슬라이싱은 값을 확인하기 위해서 사용하기 때문에 ..

wine.iloc[0:3,:]["alcohol"] = np.nan #이렇게 하면 값이 바뀌지 않는다.
```

```python
# loc를 이용하면 컬럼의 이름으로 인덱싱을 할 수 있음
wine.loc[0:3,"alcohol"] = np.nan 
```

```python
# iloc 이용하면 인덱스를 이용해서 인덱싱 할 수 있음. 결과는 위와 동일하게 나옴.
wine.iloc[0:3,0] = np.nan 
```

```python
wine
```

### Step 7. magnesium 의 3행과 4행의 값을 NaN으로 설정하세요

```python
wine.iloc[2:4,3] = np.nan
wine.head()
```
