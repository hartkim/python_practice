### Step 4. $10.00 이상인 제품이 몇 개 있습니까?
#### 같은 항목이면 비싼 물품 기준

```python
# item_price의 값들을 float으로 변경해 보세요

# 나의 코드 근데 아까는 item_price가 한번 더 들어갔는데 왜 이제는 괜찮지?
chipo["item_price"]=chipo["item_price"].str.replace(pat=r'$', repl=r'', regex=True)
```

```python
chipo.item_price[0] # '$2.39 ' 라서 $를 없애줘야한다.
```

```python
# 강사님 코드
# float으로 바꾸기
# float(chipo.item_price[0][1:])
chipo.item_price = pd.Series(map(lambda x : float(x[1:]), chipo.item_price))
chipo
```

```python
chipo[chipo.item_price >= 10].count()
# 1130개
```

### Step 5. 각 품목의 가격은 얼마입니까?
###### item_name 과 item_price 만 출력하시오

```python
one_qu=chipo[chipo.quantity ==1]
```

```python
one_qu.groupby("item_name").min().sort_values(by="item_price",ascending=False)
```

```python
# 각 품목의 가격.. 품목별 가격 이니까 group by?
a =chipo.groupby("item_name")[['item_price']].mean()
```

```python
a.sort_values(by="item_price",ascending=False)
```

### Step 9. Canned Soda를 두 개 이상 주문한 횟수는 몇번입니까?

```python
soda[soda["quantity"] >=2].count()
```

```python
# 방법 2 : 그냥 줄줄 이어쓰기
chipo[chipo.item_name == "Canned Soda"][chipo[chipo.item_name == "Canned Soda"]["quantity"] >=2].count()
```
