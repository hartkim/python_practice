### 4. 피봇 테이블과 히트맵 그리기

**문제 4:** 펭귄 종류(species)와 부리 길이(bill_length_mm)에 따른 평균 부리 깊이(bill_depth_mm)에 따른 피봇 테이블을 생성하고, 히트맵으로 시각화하세요.

```python
import pandas as pd
```

```python
data_corr=pd.pivot_table(df,                # 피벗할 데이터프레임
                     index = 'species',    # 행 위치에 들어갈 열
                     columns = 'bill_length_mm',    # 열 위치에 들어갈 열
                     values = 'bill_depth_mm',     # 데이터로 사용할 열
                     aggfunc = 'mean')
```

```python
# df.pivot_table 동작x, pd.pivot_table 동작o
pivot_df = pd.pivot_table(df, index="species",columns="bill_length_mm", values="bill_depth_mm")
sns.heatmap(pivot_df)
```
