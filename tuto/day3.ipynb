```python
score = input("점수를 입력해주세요")

if score.isdigit(): #숫자인지 아닌지 판별, 숫자면 True 문자면 False
    score = int(score) #숫자이면 int로 바꿈
    if score >=90:
        print("A")
    elif 90> score >=80:
        print("B")
    elif  80 > score >=70:
        print("C")
    else:
        print("F, 재수강하세요")
else:
    print("올바른 점수를 입력하세요")
```

```python
# 5명이 학생이 시험점수가 60점 넘으면 합격
# fn + tap 하면 한번에 tap 됨!

marks = [90,25,67,45,80]
numbers = 0

for i in marks:
    numbers = numbers+1 # numbers += 1과 같다
    if i >=60:
        print(f"{numbers}번 학생 합격입니다")
    else:
        print(f"{numbers}번 학생 불합격입니다")
```

list도 재할당을 많이 해준다.


```python
# list 재할당
list_0 = ["레드벨벳","트와이스","블랙핑크"]
list_1 = ["르세라핌","아이브","아이들","뉴진스"]
for i in list_1:
    list_0 += [i]
    print(list_0)
    print(id(list_0)) #재할당 됨. id도 동일!

```

```python
# list 추가
list_0 = ["레드벨벳","트와이스","블랙핑크"]
list_1 = ["르세라핌","아이브","아이들","뉴진스"]
for i in list_1:
    list_0.append(i)
    print(list_0)
    print(id(list_0)) 
    # append를 이용해서 추가할 수 있다.
```

```python
# 예 
# 구구단 계산기
# 몇 단을 계산하시겠습니까?
# 2x1 =2
# 2x2 =4
# ...
# 2x9 =18

a = int(input("몇 단을 계산할까요?"))
for i in range(1,10):
    hi = a*i
    print(f"{a}x{i}={hi}")
```

```python
# for 문을 이용해서 1부터 n까지의 합을 구할 수 있도록 작성
# 1부터 어디까지 더하실 겁니까?
# 55

num = int(input("어디부터 더하실 겁니까?"))
acc =0
for i in range(1, num+1):
    acc += i
print(acc)
```

```python
# 숫자 n을 입력했을 때 k번 곱하는 반복문

num = int(input("숫자를 입력하세요"))
k = int(input("몇 번 반복할까요?"))
acc = 1 #더하기가 아니고 곱하기라서 #재할당을 주의하자.
for i in range(k):
    acc *= num
print(f"{num}을{k}번 곱한 값은 {acc}입니다.")
```

```python
# for & zip

list_name = ["오민엽","김도훈","손준현","양호영"]
list_math = [11, 23, 94, 29]
list_eng = [22, 32, 11, 23]

# 3개를 동시에 리스트에 넣고 쓰고 싶다면!
for i,j,k in zip(list_name,list_math,list_eng):
    print(f"이름:{i}")
    print(f"수학점수:{j}")
    print(f"영어점수:{k}")
```

```python
# enumerate : list의 index랑 같이 출력을 해준다. 이걸 많이 쓴다.
for j,i in enumerate(list_name):
    print(f"{j}번째 이름 : {i}")
```
